```python
import re
import nltk
import string
import heapq
nltk.download('punkt')
nltk.download('stopwords')
```

# Processing Text 

```python
original_text = '''artificial intelligence is human like intelligence.
                   It is the study of intelligent artificial agents.
                   Science and engineering to produce intelligent machines.
                   Solve problems and have intelligence.
                   Related to intelligent behavior.
                   Developing of reasoning machines.
                   Learn from mistakes and successes.
                   Artificial intelligence is related to reasoning in everyday situations.'''
original_text
```

```python
original_text =re.sub(r'\s+',' ',original_text) # \s fro space and + for one or more than one space
original_text
```

```python
# chaning letter to lower case and removing stop words like is,to, the ..
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
string.punctuation
```

```python
def preprocess(text):
    formatted_text = text.lower()
    tokens = []  # storing each words of sentence with out stop words
    for token in nltk.word_tokenize(formatted_text):
        tokens.append(token)
    
    tokens = [word for word in tokens if word not in stopwords and word not in string.punctuation]
    formatted_text = ' '.join(element for element in tokens) # creating sentence without stop words and punctations
    
    return formatted_text
```

```python
formatted_text = preprocess(original_text)
print(formatted_text)
```

# Word Frequency

```python
word_frequency = nltk.FreqDist(nltk.word_tokenize(formatted_text))
for key, value in word_frequency.items():
    print(key, ' : ', value)
```

```python
len(word_frequency.keys())
```

```python
highest_freq = max(word_frequency.values())
highest_freq
```

```python
for word in word_frequency.keys():
    word_frequency[word] = (word_frequency[word] / highest_freq )
```

```python
word_frequency
```

# sentence tokentization

```python
sentence_list = nltk.sent_tokenize(original_text) # using this insted split split spereste ph.d as ph, d
sentence_list
```

# Genreating summary

```python
score_sentences = {}

for sentence in sentence_list:
    
    for word in nltk.word_tokenize(sentence.lower()):
    
        if sentence not in score_sentences.keys():
            score_sentences[sentence] = word_frequency[word]
        else:
             score_sentences[sentence] += word_frequency[word]
```

```python
score_sentences
```

```python

best_sentences = heapq.nlargest(3, score_sentences, key= score_sentences.get)
```

```python
summary = " ".join(best_sentences)
```

```python
def Summarize(text, number_of_sentences, percentage = 0):
    original_text = text
    fromatted_text = preprocess(original_text)
    
    word_frequency = nltk.FreqDist(nltk.word_tokenize(formatted_text))
    highest_freq = max(word_frequency.values())
    
    for word in word_frequency.keys():
        word_frequency[word] = (word_frequency[word] / highest_freq )
        
    sentence_list = nltk.sent_tokenize(original_text)
    
    score_sentences = {}

    for sentence in sentence_list:
    
        for word in nltk.word_tokenize(sentence.lower()):
    
            if sentence not in score_sentences.keys():
                score_sentences[sentence] = word_frequency[word]
            else:
                 score_sentences[sentence] += word_frequency[word]
                    
    if percentage > 0 :               
        best_sentences = heapq.nlargest(int(len(number_of_sentences) * percentage), score_sentences, key= score_sentences.get)
        
    else:
        best_sentences = heapq.nlargest(number_of_sentences, score_sentences, key= score_sentences.get)
        
        
    summary = " ".join(best_sentences)
    return sentence_list, best_sentences,word_frequency, score_sentences, summary  
```
